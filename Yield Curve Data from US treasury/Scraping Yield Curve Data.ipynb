## Import Libraries

```python
import pandas as pd
import requests
import random

# import numpy as np
import smtplib
from bs4 import BeautifulSoup

from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText

import time
# import sys
# from IPython.display import HTML
```

```python
# Date Range to scrap data
Year_range=range(2023,2024)
type(Year_range)
```

```python
# Date Range to scrap data
Year_range=range(2023,2024)

# Data Frame to store data
df=pd.DataFrame(columns=('Date','Header','Value'))

for year in Year_range:

    # US Tresury Yield Curve data link
    url = "https://home.treasury.gov/resource-center/data-chart-center/interest-rates/TextView?type=daily_treasury_yield_curve&field_tdr_date_value="+str(year)
    # http request
    response = requests.get(url)
    # Html parsing using beautifulsoup
    doc=BeautifulSoup(response.text,"html.parser")
    div_elements=doc.findAll('td')
    
    # iterrating each element through for loop to get data
    for x in div_elements:   
        # if Date value found store the date in temp_date variable
        if x.attrs['class'][1]=='views-field-field-tdr-date'  :
            temp_date=x.find('time')
        # Storing row data to dataframe 
        df.loc[df.shape[0]+1]=(temp_date.text,x.attrs['headers'][0],x.text)
    # random number generator to add delay between itterations
    random_number = random.randint(6, 20)
    time.sleep(random_number)  # Add a 1-second delay between iterations
# filtering date columns 
df=df[df['Header']!='view-field-tdr-date-table-column']

# Pivoting the table 
pivoted_df=df.pivot(index='Date',columns='Header',values='Value')

# Ranaming Columns
New_names=[]
for name in pivoted_df.columns.tolist():
    temp_name=name.replace('view-field-bc-','')
    temp_name=temp_name.replace('-table-column','')
    temp_name=temp_name.replace('view-field-br-round-b1','')
    temp_name=temp_name.replace('view-field-tc-','')
    temp_name=temp_name.replace('view-field-tc-','')

    New_names.append(temp_name)

pivoted_df.columns=New_names

pivoted_df.reset_index(inplace=True)

# Exporting data
pivoted_df.to_excel("Raw Data\Yield Curve.xlsx",index=False)
```
