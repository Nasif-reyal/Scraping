### Find Specific Values

```python
def extract_key_value_pairs(json_obj, key, value_filter=None):
    results = []

    if isinstance(json_obj, dict):
        for k, v in json_obj.items():
            if k == key and (value_filter is None or v == value_filter):
                results.append((k, v))  # Add key and value as a tuple
            else:
                results.extend(extract_key_value_pairs(v, key, value_filter))  # Recursive call
    elif isinstance(json_obj, list):
        for item in json_obj:
            results.extend(extract_key_value_pairs(item, key, value_filter))  # Recursive call

    return results

# Example usage
key_to_find = "frame"
value_filter = "CY2024Q1"
matches = extract_key_value_pairs(json_response, key_to_find, value_filter)
print(matches)
```

### Find path of values

```python
def extract_key_value_paths(json_obj, key, value_filter=None, path=""):
    results = []

    if isinstance(json_obj, dict):
        for k, v in json_obj.items():
            current_path = f"{path}/{k}" if path else k
            if k == key and (value_filter is None or v == value_filter):
                results.append((current_path, (k, v)))  # Include path and key-value pair
            else:
                results.extend(extract_key_value_paths(v, key, value_filter, current_path))
    elif isinstance(json_obj, list):
        for index, item in enumerate(json_obj):
            current_path = f"{path}[{index}]"
            results.extend(extract_key_value_paths(item, key, value_filter, current_path))

    return results

key_to_find = "frame"
value_filter = "CY2024Q3"
matches = extract_key_value_paths(json_response, key_to_find, value_filter)
for path, (k, v) in matches:
    print(f"Path: {path}, Key: {k}, Value: {v}")
```
